## Task 1.1 (1 point)

Let $\boldsymbol A \in\mathbb R^{m\times n}$, $\boldsymbol B \in\mathbb R^{n\times m}$. Prove that $\mathrm{tr}(\boldsymbol{AB}) = \mathrm{tr}(\boldsymbol{BA})$. Using this property, calculate $\mathrm{tr}(\boldsymbol{uv}^\mathsf{T})$ if $\boldsymbol u, \boldsymbol v \in\mathbb R^n$, $\boldsymbol u \perp \boldsymbol v$.

### YOUR SOLUTION HERE

Recall that the trace of a square matrix is the sum of its diagonal elements: 
$$ 
\mathrm{tr}(\boldsymbol{AB}) = \sum _{i=1} ^m \boldsymbol{(AB)}_{ii}
$$
i-th diagonal entry of $\boldsymbol{(AB)_{ii}}$ is given by:
$$
\boldsymbol{(AB)_{ii}} = \sum _{k=1} ^n \boldsymbol{A_{ik}B_{ki}}
$$
Thus : $$\mathrm{tr}(\boldsymbol{AB}) = \sum _{i=1} ^m \sum _{k=1} ^n \boldsymbol{A_{ik}B_{ki}}$$
After the same steps with $\mathrm{tr}(\boldsymbol{BA})$:
$$
\mathrm{tr}(\boldsymbol{BA}) = \sum _{j=1} ^n \boldsymbol{(BA)_{jj}} = 
\sum _{j=1} ^n \sum _{l=1} ^m \boldsymbol{B_{jl}A_{lj}}
\space (j = k, l = i)= \sum _{k=1} ^n \sum _{i=1} ^m \boldsymbol{B_{ki}A_{ik}}
$$

Solution for $\mathrm{tr}(\boldsymbol{uv}^\mathsf{T})$ when $\boldsymbol u, \boldsymbol v \in\mathbb R^n$, $\boldsymbol u \perp \boldsymbol v$:
$$
\mathrm{tr}(\boldsymbol{uv}^\mathsf{T}) = \sum _{i=1} ^n \boldsymbol{(uv^\mathsf{T})}_{ii}
$$
where: $  \enspace\hspace{6cm}   \boldsymbol{(uv^\mathsf{T})}_{ii} = u_iv_i$

Thus: $$\mathrm{tr}(\boldsymbol{uv}^\mathsf{T}) = \sum _{i=1} ^n u_iv_i = u^\mathsf{T}v$$

Since $\boldsymbol u \perp \boldsymbol v \to u^\mathsf{T}v = 0$.

## Task 1.2 (1 point)

Let $\boldsymbol A \in \mathbb R^{m\times n}$. Prove that $N(\boldsymbol A) = N(\boldsymbol A^{\mathsf T}\boldsymbol A)$, i.e., matrices $\boldsymbol A$ and $\boldsymbol A^{\mathsf T} \boldsymbol A$ have same nullspaces.

### YOUR SOLUTION HERE
Proving that the nullspace of a matrix $\boldsymbol{A} \in \mathbb{R}^{m \times n}$ is the same as the nullspace of $\boldsymbol{A}^\mathsf{T} \boldsymbol{A}$, i.e.,  
$$ 
N(\boldsymbol{A}) = N(\boldsymbol{A}^\mathsf{T} \boldsymbol{A}). 
$$ 
 
Step 1: Nullspace of $\boldsymbol{A}$ 
 
By definition, the nullspace of $\boldsymbol{A}$, denoted as $N(\boldsymbol{A})$, is the set of vectors $\boldsymbol{x} \in \mathbb{R}^n$ such that: 
$$ 
\boldsymbol{A}\boldsymbol{x} = 0. 
$$ 
In other words, 
$$ 
N(\boldsymbol{A}) = \{ \boldsymbol{x} \in \mathbb{R}^n \mid \boldsymbol{A} \boldsymbol{x} = 0 \}. 
$$ 
 
Step 2: Nullspace of $\boldsymbol{A}^\mathsf{T} \boldsymbol{A}$ 
 
Similarly, the nullspace of $\boldsymbol{A}^\mathsf{T} \boldsymbol{A}$, denoted as $N(\boldsymbol{A}^\mathsf{T} \boldsymbol{A})$, is the set of vectors $\boldsymbol{x} \in \mathbb{R}^n$ such that: 
$$ 
\boldsymbol{A}^\mathsf{T} \boldsymbol{A} \boldsymbol{x} = 0. 
$$ 
In other words, 
$$ 
N(\boldsymbol{A}^\mathsf{T} \boldsymbol{A}) = \{ \boldsymbol{x} \in \mathbb{R}^n \mid \boldsymbol{A}^\mathsf{T} \boldsymbol{A} \boldsymbol{x} = 0 \}. 
$$ 
 
Step 3: Show that $N(\boldsymbol{A}) \subseteq N(\boldsymbol{A}^\mathsf{T} \boldsymbol{A})$ 
 
Let $\boldsymbol{x} \in N(\boldsymbol{A})$. This means: 
$$ 
\boldsymbol{A} \boldsymbol{x} = 0. 
$$ 
Now, we need to check if $\boldsymbol{x} \in N(\boldsymbol{A}^\mathsf{T} \boldsymbol{A})$. Consider: 
$$ 
\boldsymbol{A}^\mathsf{T} \boldsymbol{A} \boldsymbol{x} = \boldsymbol{A}^\mathsf{T} (\boldsymbol{A} \boldsymbol{x}). 
$$ 
Since $\boldsymbol{A} \boldsymbol{x} = 0$, we have: 
$$ 
\boldsymbol{A}^\mathsf{T} \boldsymbol{A} \boldsymbol{x} = \boldsymbol{A}^\mathsf{T} 0 = 0. 
$$ 
Thus, $\boldsymbol{x} \in N(\boldsymbol{A}^\mathsf{T} \boldsymbol{A})$. 
 
Therefore, we have shown that: 
$$ 
N(\boldsymbol{A}) \subseteq N(\boldsymbol{A}^\mathsf{T} \boldsymbol{A}). 
$$ 
 
Step 4: Show that $N(\boldsymbol{A}^\mathsf{T} \boldsymbol{A}) \subseteq N(\boldsymbol{A})$ 
 
Let $\boldsymbol{x} \in N(\boldsymbol{A}^\mathsf{T} \boldsymbol{A})$. This means: 
$$ 
\boldsymbol{A}^\mathsf{T} \boldsymbol{A} \boldsymbol{x} = 0. 
$$ 
We need to show that $\boldsymbol{A} \boldsymbol{x} = 0$. 
 
Consider the expression $\boldsymbol{A}^\mathsf{T} \boldsymbol{A} \boldsymbol{x} = 0$. This implies that: 
$$ 
\langle \boldsymbol{A} \boldsymbol{x}, \boldsymbol{A} \boldsymbol{x} \rangle = (\boldsymbol{A} \boldsymbol{x})^\mathsf{T} (\boldsymbol{A} \boldsymbol{x}) = 0, 
$$ 
where $\langle \cdot, \cdot \rangle$ is the standard inner product in $\mathbb{R}^m$. 
 
Since the inner product of $\boldsymbol{A} \boldsymbol{x}$ with itself is zero, it follows that $\boldsymbol{A} \boldsymbol{x} = 0$. 
 
Thus, $\boldsymbol{x} \in N(\boldsymbol{A})$, and we have shown that: 
$$ 
N(\boldsymbol{A}^\mathsf{T} \boldsymbol{A}) \subseteq N(\boldsymbol{A}). 
$$ 
 
Step 5: Conclusion 
 
Since we have shown both: 
$$ 
N(\boldsymbol{A}) \subseteq N(\boldsymbol{A}^\mathsf{T} \boldsymbol{A}) \quad \text{and} \quad N(\boldsymbol{A}^\mathsf{T} \boldsymbol{A}) \subseteq N(\boldsymbol{A}), 
$$ 
it follows that: 
$$ 
N(\boldsymbol{A}) = N(\boldsymbol{A}^\mathsf{T} \boldsymbol{A}). 
$$ 
 
This completes the proof.

Show that similar matrices have equal determinants: $\det \boldsymbol A = \det \boldsymbol B$ if $\boldsymbol A \sim \boldsymbol B$.

### YOUR SOLUTION HERE

Two matrices $\boldsymbol{A}$ and $\boldsymbol{B}$ are said to be similar if there exists an invertible matrix $\boldsymbol{P}$ such that:
$$A = PBP^{-1}$$
Thus:
$$\det \boldsymbol A = \det \boldsymbol (PBP^{-1})$$

Apply one of the properties of determinant: 
$\det \boldsymbol (PBP^{-1}) = \det \boldsymbol(P) \det \boldsymbol(B)
\det \boldsymbol(P^{-1}).\space$ Another propertie is: $\det \boldsymbol(A^{-1}) = \frac{1}{\det \boldsymbol({A})}\newline \newline$
 Thus:
$$
\det \boldsymbol(P) \det \boldsymbol(B)\det \boldsymbol(P^{-1}) = \det \boldsymbol(P) \det \boldsymbol(B)
\frac{1}{\det \boldsymbol({P})} = \det \boldsymbol(B).
$$
From the equation above:
$\det \boldsymbol (A) = \det \boldsymbol(P) \det \boldsymbol(B)\det \boldsymbol(P^{-1}) = \det \boldsymbol(B)$

## Task 1.4 (Sherman—Morrison formula, 1.5 points)

Let $\boldsymbol A \in \mathbb R^{n\times n}$ be an invertible matrix, $\boldsymbol u, \boldsymbol v \in \mathbb R^n$ and $\boldsymbol v^\mathsf{T}\boldsymbol A^{-1} \boldsymbol u \ne -1$ . Prove that

$$
    (\boldsymbol A + \boldsymbol u \boldsymbol v^\mathsf{T})^{-1} =
    \boldsymbol A^{-1} - \frac{\boldsymbol A^{-1} \boldsymbol u \boldsymbol v^\mathsf{T} \boldsymbol A^{-1}}{1 + \boldsymbol v^\mathsf{T}\boldsymbol A^{-1} \boldsymbol u }.
$$

### YOUR SOLUTION HERE
We are tasked with proving the Sherman–Morrison formula, which states that for an invertible matrix $\boldsymbol{A} \in \mathbb{R}^{n \times n}$, vectors $\boldsymbol{u}, \boldsymbol{v} \in \mathbb{R}^n$, and the condition that $\boldsymbol{v}^\mathsf{T} \boldsymbol{A}^{-1} \boldsymbol{u} \neq -1$, the following identity holds: 
 
$$ 
(\boldsymbol{A} + \boldsymbol{u} \boldsymbol{v}^\mathsf{T})^{-1} = \boldsymbol{A}^{-1} - \frac{\boldsymbol{A}^{-1} \boldsymbol{u} \boldsymbol{v}^\mathsf{T} \boldsymbol{A}^{-1}}{1 + \boldsymbol{v}^\mathsf{T} \boldsymbol{A}^{-1} \boldsymbol{u}}. 
$$ 
 
Step 1: Assume the inverse exists 
 
Let $\boldsymbol{B} = \boldsymbol{A} + \boldsymbol{u} \boldsymbol{v}^\mathsf{T}$. We want to express $\boldsymbol{B}^{-1}$ in terms of $\boldsymbol{A}^{-1}$ and the vectors $\boldsymbol{u}, \boldsymbol{v}$. 
 
Step 2: Apply the Woodbury matrix identity 
 
Using the Woodbury matrix identity for a rank-one update, we have: 
$$ 
(\boldsymbol{A} + \boldsymbol{u} \boldsymbol{v}^\mathsf{T})^{-1} = \boldsymbol{A}^{-1} - \boldsymbol{A}^{-1} \boldsymbol{u} \frac{1}{1 + \boldsymbol{v}^\mathsf{T} \boldsymbol{A}^{-1} \boldsymbol{u}} \boldsymbol{v}^\mathsf{T} \boldsymbol{A}^{-1}. 
$$ 
 
Step 3: Simplify the expression 
 
Since $\frac{1}{1 + \boldsymbol{v}^\mathsf{T} \boldsymbol{A}^{-1} \boldsymbol{u}}$ is a scalar, it commutes with all matrix terms. Therefore, we can write: 
$$ 
(\boldsymbol{A} + \boldsymbol{u} \boldsymbol{v}^\mathsf{T})^{-1} = \boldsymbol{A}^{-1} - \frac{\boldsymbol{A}^{-1} \boldsymbol{u} \boldsymbol{v}^\mathsf{T} \boldsymbol{A}^{-1}}{1 + \boldsymbol{v}^\mathsf{T} \boldsymbol{A}^{-1} \boldsymbol{u}}. 
$$ 
 
Step 4: Conclusion 
 
Thus, we have proven that for an invertible matrix $\boldsymbol{A}$ and vectors $\boldsymbol{u}, \boldsymbol{v} \in \mathbb{R}^n$, provided that $\boldsymbol{v}^\mathsf{T} \boldsymbol{A}^{-1} \boldsymbol{u} \neq -1$, the Sherman–Morrison formula holds: 
$$ 
(\boldsymbol{A} + \boldsymbol{u} \boldsymbol{v}^\mathsf{T})^{-1} = \boldsymbol{A}^{-1} - \frac{\boldsymbol{A}^{-1} \boldsymbol{u} \boldsymbol{v}^\mathsf{T} \boldsymbol{A}^{-1}}{1 + \boldsymbol{v}^\mathsf{T} \boldsymbol{A}^{-1} \boldsymbol{u}}. 
$$

## Task 1.5 (programming, 1.5 points)

Apply k-NN algorithm to MNIST dataset and measure its performance:

* split the datasert into train and test parts 
* train several models with different hyperparameters (take $1\leqslant k \leqslant 20$ and different distance metrics ($p=1$, $p=2$, $p=+\infty$))
* visualize several test samples and their predictions (see code below)
* plot train and test accuracies for each model on the same graph
* find a model with best test accuracy

Fetch MNIST dataset:

```python
from sklearn.datasets import fetch_openml
X, Y = fetch_openml('mnist_784', return_X_y=True, parser='auto')
X = X.astype(float).values / 255
Y = Y.astype(int).values  # type: ignore
X.shape, Y.shape
```

Visualize it:

```python
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'svg'

def plot_digits(X, y_true, y_pred=None, n=4, random_state=123):
    np.random.seed(random_state)
    indices = np.random.choice(np.arange(X.shape[0]), size=n*n, replace=False)
    plt.figure(figsize=(10, 10))
    for i in range(n*n):
        plt.subplot(n, n, i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(X[indices[i]].reshape(28, 28), cmap='gray')
        # plt.imshow(train_images[i], cmap=plt.cm.binary)
        if y_pred is None:
            title = str(y_true[indices[i]])
        else:
            title = f"y={y_true[indices[i]]}, ŷ={y_pred[indices[i]]}"
        plt.title(title, size=20)
    plt.show()

plot_digits(X, Y)
```

```python
# YOUR CODE HERE
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

k_values = range(3, 8)  # k from 1 to 20
p_values = [1, 2, 7]  # p=1 (Manhattan), p=2 (Euclidean), p=∞ (Chebyshev)

# Store accuracies
train_accuracies = []
test_accuracies = []

# Train models with different k and p values
for p in p_values:
    train_acc = []
    test_acc = []
    for k in k_values:
        # Initialize k-NN classifier with parallel processing
        knn = KNeighborsClassifier(n_neighbors=k, p=p, n_jobs=-1)  # Use all CPU cores
        
        # Train the classifier
        knn.fit(X_train, y_train)
        
        # Training accuracy
        y_train_pred = knn.predict(X_train)
        train_acc.append(accuracy_score(y_train, y_train_pred))
        
        # Test accuracy
        y_test_pred = knn.predict(X_test)
        test_acc.append(accuracy_score(y_test, y_test_pred))
    
    # Store the results for each p
    train_accuracies.append(train_acc)
    test_accuracies.append(test_acc)
    
    print(f"Done training for p = {p}")
```

```python
y_test_pred = knn.predict(X_test)
plot_digits(X_test, y_test, y_pred=y_test_pred)
```

```python
import matplotlib.pyplot as plt

# Plot the results
plt.figure(figsize=(10, 6))

for i, p in enumerate(p_values):
    plt.plot(k_values, train_accuracies[i], label=f'Train Accuracy (p={p})', linestyle='--')
    plt.plot(k_values, test_accuracies[i], label=f'Test Accuracy (p={p})', linestyle='-')

plt.xlabel('k (Number of Neighbors)')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Train and Test Accuracies for Different k and p values')
plt.show()
```

```python
best_accuracy = 0
best_k = 0
best_p = 0

for i, p in enumerate(p_values):
    for j, k in enumerate(k_values):
        if test_accuracies[i][j] > best_accuracy:
            best_accuracy = test_accuracies[i][j]
            best_k = k
            best_p = p

print(f"The best model has k={best_k}, p={best_p} with a test accuracy of {best_accuracy:.4f}")
```

## Task 1.6 (programming, 1.5 points)

Compare the performance of different algorithms for calculation matrix product 

$$
       \boldsymbol C = \boldsymbol {AB}, \quad \boldsymbol A \in \mathbb R^{m\times n}, \quad \boldsymbol B \in \mathbb R^{n\times p}.
$$

Try the following methods:

* nested pythonic loop implementing the formula $C_{ik} = \sum\limits_{j=1}^n A_{ij}B_{jk}$
* replace the inner `for` loop by `np.dot` for calculating $\boldsymbol a_i^\mathsf{T}\boldsymbol b_k$
* calculate $\sum\limits_{j=1}^n \boldsymbol a_j \boldsymbol b_j^\mathsf{T}$ (use [np.outer](https://numpy.org/doc/stable/reference/generated/numpy.outer.html))
* just use numpy and calculate `A @ B`

The plan (see aslo dot product demo below):

* implement these four functions
* test that they return the same result
* measure their performance on a square matrix of shape $100\times 100$
* plot graphs of the execution time versus $n$

```python
import numpy as np

def mat_mul_loop(A, B: np.array) -> np.array:
    # YOUR CODE HERE
    n, m = A.shape
    m, p = B.shape
    C = np.zeros((n, p))
    for i in range(n):
        for j in range(p):
            for k in range(m):
                C[i, j] += A[i, k] * B[k, j]
    return C

def mat_mul_inner(A, B: np.array) -> np.array:
    # YOUR CODE HERE
    n, m = A.shape
    m, p = B.shape
    C = np.zeros((n, p))
    for i in range(n):
        C[i, :] = np.dot(A[i, :], B)
    return C
 

def mat_mul_outer(A, B: np.array) -> np.array:
    # YOUR CODE HERE
    n, m = A.shape
    m, p = B.shape
    C = np.zeros((n, p))
    for i in range(m):
        C += np.outer(A[:, i], B[i, :])
    return C

def mat_mul_np(A, B: np.array) -> np.array:
    # YOUR CODE HERE
    return A @ B
    
```

```python
# TESTING AREA

def TestMatrices(n_max):
    for n in range(1, n_max):
        m = np.random.randint(1, n_max)
        p = np.random.randint(1, n_max)
        A = np.random.randn(m, n)
        B = np.random.randn(n, p)
        prod_loop = mat_mul_loop(A, B)
        prod_inner = mat_mul_inner(A, B)
        prod_outer = mat_mul_outer(A, B)
        prod_np = mat_mul_np(A, B)
        assert np.allclose(prod_loop, prod_inner)
        assert np.allclose(prod_loop, prod_np)
        assert np.allclose(prod_loop, prod_outer)

TestMatrices(100)
```

```python
import matplotlib.pyplot as plt
from time import time

%config InlineBackend.figure_format = 'svg'

def measure_time(func, m, n, p, n_samples=10):
    result = np.zeros(n_samples)
    for i in range(n_samples):
        begin = time()
        func(np.random.randn(m, n), np.random.randn(n, p))
        result[i] = time() - begin
    return result.mean()

def get_times_lists(func, step=10, max_size=200, n_samples=20):
    times = []
    sizes = np.arange(step, max_size + 1, step)
    for size in sizes:
        times.append(measure_time(func, size, size, size, n_samples))
    return np.array(times)

def plot_time_vs_size(step=10, max_size=100, n_samples=20):
    loop_times = 1000*get_times_lists(mat_mul_loop, step, max_size, n_samples)
    inner_times = 1000*get_times_lists(mat_mul_inner, step, max_size, n_samples)
    outer_times = 1000*get_times_lists(mat_mul_outer, step, max_size, n_samples)
    np_times = 1000*get_times_lists(mat_mul_np, step, max_size, n_samples)
    sizes = np.arange(step, max_size + 1, step)
    plt.semilogy(sizes, loop_times, c='r', lw=2, label="loop")
    plt.semilogy(sizes, inner_times, c='b', lw=2, label="inner")
    plt.semilogy(sizes, outer_times, c='g', lw=2, label="outer")
    plt.semilogy(sizes, np_times, c='m', lw=2, label="np")
    plt.xlim(0, max_size)
    plt.title("Matrix product")
    plt.legend()
    plt.xlabel("size")
    plt.ylabel("time, ms")
    plt.grid(ls=":");
```

## Demo for task 1.6: dot product performance

Compare several implementations of the dot product.

```python
import numpy as np

def dot_loop(a, b):
    result = 0
    for i in range(len(a)):
        result += a[i] * b[i]
    return result

def dot_sum(a, b):
    return sum(a * b)

def dot_np(a, b):
    return a @ b
```

Test that all this functions produce the same result:

```python
for n in range(1, 101):
    a = np.random.randn(n)
    b = np.random.randn(n)
    assert np.allclose(dot_loop(a, b), dot_sum(a, b))
    assert np.allclose(dot_loop(a, b), dot_np(a, b))
    assert np.allclose(dot_np(a, b), dot_sum(a, b))
```

This is why you should almost always prefer `numpy` to pythonic loops! To emphasize this effect, plot the graphs of execution time versus array size.

```python
import matplotlib.pyplot as plt
from time import time

%config InlineBackend.figure_format = 'svg'

def measure_time(func, size, n_samples=10):
    result = np.zeros(n_samples)
    for i in range(n_samples):
        begin = time()
        func(np.random.randn(size), np.random.randn(size))
        result[i] = time() - begin
    return result.mean()

def get_times_lists(func, step=20, max_size=1000, n_samples=20):
    times = []
    sizes = np.arange(20, max_size + 1, 20)
    for size in sizes:
        times.append(measure_time(func, size, n_samples))
    return np.array(times)

def plot_time_vs_size(step=20, max_size=1000, n_samples=50):
    loop_times = 1000*get_times_lists(dot_loop, step, max_size, n_samples)
    sum_times = 1000*get_times_lists(dot_sum, step, max_size, n_samples)
    np_times = 1000*get_times_lists(dot_np, step, max_size, n_samples)
    sizes = np.arange(step, max_size + 1, step)
    plt.plot(sizes, loop_times, c='r', lw=2, label="loop")
    plt.plot(sizes, sum_times, c='b', lw=2, label="sum")
    plt.plot(sizes, np_times, c='g', lw=2, label="np")
    plt.xlim(0, max_size)
    plt.title("Dot product")
    plt.legend()
    plt.xlabel("size")
    plt.ylabel("time, ms")
    plt.grid(ls=":");
```
